```python
from great_tables import GT, style, loc
import polars as pl
import polars.selectors as cs
from datetime import datetime

MIN_YEAR = 1948
MAX_YEAR = datetime.today().year

df = (
    pl.read_csv('../unemployment_rate.csv')
    .filter(pl.col('Year')!=MAX_YEAR)
    .transpose(
        include_header=True, 
        header_name='Month',
        column_names=[str(num) for num in range(MIN_YEAR, MAX_YEAR)]
    )
    .slice(1,-1)
    .select(pl.col('Month'), cs.starts_with('20'))
)

(
    GT(df, rowname_col='Month')
    .data_color(
        domain=[0, 15],
        palette=['#eaf0fa', '#d5e1f5', '#c0d2f0', '#abc3eb', '#96b4e6', '#81a5e1', '#6c96dc', '#5787d7', '#4278d2', '#2e69ce'],
        na_color='white',
    )
    .tab_header(
        title=f'US Unemployment Rate: {2020} ~ {MAX_YEAR-1}',
    )
)
```
